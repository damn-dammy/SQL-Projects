```python
# what are the top 25 industries as related to revenue and profits
df.groupby('Industry')[['Revenues', 'Profits']].sum().sort_values(by='Revenues', ascending=False).head(25)
Top_25_industries = df.groupby('Industry')[['Revenues', 'Profits']].sum().sort_values(by='Revenues', ascending=False).head(25)
print (Top_25_industries)

```

```python
# Convert 'Revenues' column back to numeric before applying formatting
Top_25_industries['Revenues'] = pd.to_numeric(Top_25_industries['Revenues'].str.replace('$|,', '', regex=True))
#Now apply the desired formatting
Top_25_industries['Revenues'] = Top_25_industries['Revenues'].apply(lambda x: '${:,.2f}'.format(x))
print(Top_25_industries)
```

```python
# lowest 25 companies as it relates to revenue and profits.
df.groupby('Industry')[['Revenues', 'Profits']].sum().sort_values(by='Revenues', ascending=True).head(25)
```

```python
# what industries have the most assets
df.groupby('Industry')['Assets'].sum().sort_values(ascending=False).head(25)
```

```python
# what industries have the least assets
df.groupby('Industry')['Assets'].sum().sort_values(ascending=True).head(25)
```

```python
#whats the percentage of profits across all industries
Percentage_of_industry_revenue = df.groupby('Industry')['Profits'].sum() / df['Profits'].sum() * 100
Top_10_profit_industries = Percentage_of_industry_revenue.sort_values(ascending=False).head(10)
Top_10_profit_industries
```

```python
plt.figure(figsize=(9,9))
plt.pie(Top_10_profit_industries, labels=Top_10_profit_industries.index, shadow=True, startangle=140, autopct= '%1.1f%%')
plt.title ('Percentage of Profits by Industry')
plt.tight_layout()
plt.show()
```

```python
industry_revenue = df.groupby('Industry')['Revenues'].sum().sort_values(ascending=False).head(10)
industry_revenue  = industry_revenue.reset_index()
industry_revenue.columns = ['Industry', 'Total Revenue']
industry_revenue
```

```python
from os import pardir
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='Industry', y='Total Revenue', data=industry_revenue,)
plt.title('Top 10 Industries by Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()
```

```python
#3 Calculate profit margins
df['Profit Margin (%)'] = (df['Profits'] / df['Revenues']) * 100
top_margin_companies = df[['Title', 'Profit Margin (%)']].sort_values(by='Profit Margin (%)', ascending=False).head(10)
top_margin_companies
```

```python
#3 what companies have negative  profits
loss_making_companies = df[df['Profits'] < 0][['Title', 'Profits']].sort_values(by='Profits', ascending=True).head(10)
loss_making_companies
```

```python
sns.barplot(x='Profits', y='Title', data=loss_making_companies)
plt.title('Top 10 Companies with Negative Profits')
plt.xlabel('Profits')
plt.ylabel('Company')
plt.tight_layout()
plt.show()
```

```python
# What is the correlation between revenue and profit?
correlation = df[['Revenues', 'Profits']].corr()
correlation
```

```python
#3 Relationship between revenue and Profit.
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Revenues'], y=df['Profits'], alpha=0.7)
plt.xlabel("Revenue (in billions)")
plt.ylabel("Profit (in billions)")
plt.title("Revenue vs. Profit of Fortune 500 Companies")
plt.show()
```

```python
#3 Revenue Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Revenues'], bins=30, kde=True, color='blue')
plt.xlabel("Revenue (in billions)")
plt.ylabel("Number of Companies")
plt.title("Revenue Distribution of Fortune 500 Companies")
plt.show()
```
